```python
from sklearn.datasets import fetch_mldata
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
mnist = fetch_mldata('MNIST original')
X,y = mnist["data"], mnist["target"]
```

```python
unit_digit = X[66000]
unit_digit_image = unit_digit.reshape(28,28)
```

```python
y[66000]
```

Dimensionality Reduction

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 154)
x_mnist_reduced = pca.fit_transform(X)
```

```python
np.shape(x_mnist_reduced)
```

```python
inversed_img = pca.inverse_transform(x_mnist_reduced)
```

Output of original images

```python
fig, axes = plt.subplots(1,6,figsize=(24,4),
                        subplot_kw={'xticks':[], 'yticks':[]},
                        gridspec_kw=dict(hspace=0.1, wspace=0.1))

a = [10000,20000,30000,50000,60000,68000]

for i, ax in enumerate(axes.flat):
    ax.imshow(X[a[i]].reshape(28,28), cmap = matplotlib.cm.binary, interpolation="nearest")
```

Output of inversed images

```python
fig, axes = plt.subplots(1,6,figsize=(24,4),
                        subplot_kw={'xticks':[], 'yticks':[]},
                        gridspec_kw=dict(hspace=0.1, wspace=0.1))

a = [10000,20000,30000,50000,60000,68000]

for i, ax in enumerate(axes.flat):
    ax.imshow(inversed_img[a[i]].reshape(28,28), cmap = matplotlib.cm.binary, interpolation="nearest")
```

Printing features

```python
fig, axes = plt.subplots(1,8,figsize=(30,4),
                        subplot_kw={'xticks':[], 'yticks':[]},
                        gridspec_kw=dict(hspace=0.1, wspace=0.1))

for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(28,28), cmap='bone')
```

Reducing the number of bits

```python
x_mnist_compressed = np.round(x_mnist_reduced * 4, 0) / 4.0 
inversed_compressed_img = pca.inverse_transform(x_mnist_compressed)


fig, axes = plt.subplots(1,6,figsize=(30,4),
                        subplot_kw={'xticks':[], 'yticks':[]},
                        gridspec_kw=dict(hspace=0.1, wspace=0.1))

a = [10000,20000,30000,50000,60000,68000]

for i, ax in enumerate(axes.flat):
    ax.imshow(inversed_compressed_img[a[i]].reshape(28,28), cmap = matplotlib.cm.binary, interpolation="nearest")

```

<b>Discussion</b><br>
    <p>The dimensionality algorithm PCA does a formidable job without losing imported features of the images. Another significant observation is that the images were almost identical to the dimensionality reduced images even if the values were compressed (rounded-off)</p>
